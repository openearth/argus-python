```python
# get argus grid
r = argus.rest.get_rectification_data('kijkduin')
H = flamingo.rectification.find_homography(r[cid]['UV'], r[cid]['XYZ'], r[cid]['K'])
u, v = flamingo.rectification.get_pixel_coordinates(img)
u, v = argus.plot.undistort(u, v, rectification_data=r[cid])
X, Y = flamingo.rectification.rectify_coordinates(u, v, H)

# get real-world grid
r = argus.rest.get_rotation_data('kijkduin')
o = flamingo.rectification.plot.find_horizon_offset(X, Y, max_distance=1000.)
Xr, Yr = flamingo.rectification.plot.rotate_translate(-X[o:,:], Y[o:,:], **r)

# get bathy data
nc = netCDF4.Dataset('https://bashoonhout:Rf3QyUs4@zandmotordata.nl/thredds/dodsC/opendap/morphology/NeMo/morphology_transects_nemo.nc', 'r')
t = netCDF4.num2date(nc.variables['time'][7], nc.variables['time'].units)
zmx = nc.variables['x'][:,:]
zmy = nc.variables['y'][:,:]
zmz = nc.variables['altitude'][7,:,:]
nc.close()

# interpolate bathy to real-world grid
Zr = scipy.interpolate.griddata(np.asarray([zmx.flatten(), zmy.flatten()]).T,
                                zmz.flatten(),
                                np.asarray([Xr.flatten(), Yr.flatten()]).T)
Z = Zr.reshape(Xr.shape)
```

```python
dts = []
dst = []
hgt = []
for dt, y in zip(ts, ys):
    if len(y)>0 and ~np.all(np.isnan(y)):

        # fill nan's
        y[np.isnan(y)] = int(np.median(y[~np.isnan(y)]))
        
        dts.append(dt)
        dst.append(Y[y.astype(int),  range(Y.shape[1])])
        hgt.append(Z[y.astype(int)-o,range(Y.shape[1])])
        
hgt = np.asarray(hgt)
dst = np.asarray(dst)
```

```python
df = pd.DataFrame({'min':np.min(hgt, axis=1),
                   'max':np.max(hgt, axis=1),
                   'std':np.std(hgt, axis=1),
                   'mean':np.mean(hgt, axis=1),
                   'median':np.median(hgt, axis=1),
                   'mid':hgt[:,1224]},
                   index=dts)
df
```
