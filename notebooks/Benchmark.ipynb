```python
import time
import flamingo.classification as cls
import flamingo.segmentation as seg
```

```python
from itertools import product

import numpy as np
import skimage.color

_FREQS = np.arange(0.05, 0.30, 0.1)
_SIGMAS = [3, 5, 7]
_SIGMA1 = range(1, 21, 7)
_THETAS = np.arange(0, np.pi, 0.25*np.pi)
```

```python
colorspace = 'rgb'
channelstats = None

if colorspace.lower() == 'rgb':
    gray = img.mean(-1)
elif colorspace.lower() == 'hsv':
    gray = img[..., -1]
else:
    raise ValueError(
        'Unsupported colorspace [%s]' % colorspace)

# create extra channels                                                                                                                                                          
channels = []

# Create a set of gabor kernels and apply them                                                                                                                                   
t0 = time.time()
for i,(frequency, sigma) in enumerate(product(_FREQS,_SIGMAS)):
    response = []
    for theta in _THETAS:
        t1 = time.time()
        real, imag = skimage.filter.gabor_filter(
            gray,
            frequency=frequency,
            sigma_x=sigma,
            sigma_y=sigma,
            theta=theta
            )
        print('    %s seconds    [f=%f, s=%f, t=%f]' % (time.time() - t1, frequency, sigma, theta))
        response.append(np.sqrt(real ** 2 + imag ** 2))

    response = np.array(response).max(axis=0)
    
    if channelstats:
        response = _normalize_channel(response, channelstats[i])
    channels.append(response)
print('--- %s seconds ---' % (time.time() - t0))
```

```python
# Create relative greyishness                                                                                                                                                    
t0 = time.time()
for i,sigma1 in enumerate(_SIGMA1):
    t1 = time.time()
    # blur                                                                                                                                                                       
    response1 = skimage.filter.gaussian_filter(gray, sigma1)
    # blur on a higher level                                                                                                                                                     
    response2 = skimage.filter.gaussian_filter(gray, sigma1 + 1)
    # create relative darkness                                                                                                                                                   
    response = response2 - response1
    print('    %s seconds    ' % (time.time() - t1))
    if channelstats:
        response = _normalize_channel(response,channelstats[i + len(_FREQS)*len(_SIGMAS)])
    channels.append(response)
print('--- %s seconds ---' % (time.time() - t0))
```

```python
t0 = time.time()
# Combine all channels                                                                                                                                                           
allchannels = np.concatenate([
    # greyscale                                                                                                                                                                  
    np.uint8(gray)[..., np.newaxis],
    # color                                                                                                                                                                      
    img
] + [
    # other channels                                                                                                                                                             
    channel[..., np.newaxis]
    for channel
    in channels
], -1)
print('--- %s seconds ---' % (time.time() - t0))
```

```python
t0 = time.time()
segments_orig = seg.superpixels.get_superpixel(
    img, n_segments=600, compactness=20, sigma=0)
print('--- %s seconds ---' % (time.time() - t0))

# compute superpixel grid   
t0 = time.time()
nx, ny = seg.superpixels.get_superpixel_grid(segments_orig, img.shape[:2])
print('--- %s seconds ---' % (time.time() - t0))

# fix disjoint segments  
t0 = time.time()
segments = seg.postprocess.remove_disjoint(segments_orig)
print('--- %s seconds ---' % (time.time() - t0))

# get contours
t0 = time.time()
contours = seg.superpixels.get_contours(segments)
print('--- %s seconds ---' % (time.time() - t0))
```

```python
features = []
features0 = []
features_in_block = {}

blocks = cls.features.blocks.list_blocks()

done = []
t0 = time.time()
while True:
    n = len(done)
    for block, fcn in blocks.iteritems():
            if not block in done:
                try:
                    t1 = time.time()
                    features = fcn(allchannels,
                                   segments,
                                   features=features,
                                   colorspace='rgb')
                    print('    %s seconds    [%s]' % (time.time() - t1, block))

                    features_in_block[block] = [
                        x for x in features.columns if x not in features0]
                    features0 = features.columns
    
                    done.append(block)
                except ValueError as e:
                    continue
    
    if len(done) == len(blocks):
            break

    if len(done) == n:
        # nothing done, raise error                                                                                                                                              
        raise StopIteration('Cannot determine order of blocks')
print('--- %s seconds ---' % (time.time() - t0))

t0 = time.time()
features = cls.features.remove_large_features(features)
print('--- %s seconds ---' % (time.time() - t0))
```
