```python
import matplotlib.pyplot as pl
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
beachcoor = np.array([35,35,10])
seacoor = np.array([20,20,50])
bpic = [np.random.normal(beachcoor[0],8,200),np.random.normal(beachcoor[1],6,200)]
spic = [np.random.normal(seacoor[0],6,200),np.random.normal(seacoor[1],7,200)]
```

```python
fig = pl.figure(figsize=(7,7))
ax = fig.add_subplot(111)
p1 = pl.scatter(bpic[0],bpic[1],color='blue')#,alpha=.1)
p2 = pl.scatter(spic[0],spic[1],color='red')#,alpha=.1)
pl.xlim((0,60))
pl.ylim((0,60))
pl.legend(['Beach','Sea'])
pl.xlabel('Hue')
pl.ylabel('Saturation')
ax.set_aspect('equal')
pl.savefig('2D.png')
pl.scatter(beachcoor[0],beachcoor[1],color='b',s=500,marker='*')
pl.scatter(seacoor[0],seacoor[1],color='r',s=500,marker='*')
p1.set_alpha(.1)
p2.set_alpha(.1)
pl.savefig('2D_center.png')
meancoor = np.mean([beachcoor[0:2],seacoor[0:2]],0)
slope = (seacoor[1]-beachcoor[1])/(seacoor[0]-beachcoor[0])
pl.plot(meancoor+np.array([-100,100]),meancoor-np.array([-100,100]),color='black')
pl.savefig('2D_line.png')
#ax = fig.gca
pl.scatter(20,50,s=50,color='g')
pl.savefig('2D_new.png')
pl.arrow(20,50,beachcoor[0]-20,beachcoor[1]-50,color='black',shape='full', lw=2, length_includes_head=True, head_width=1.5)
p3 = pl.arrow(20,50,seacoor[0]-20,seacoor[1]-50,color='black',shape='full', lw=2, length_includes_head=True, head_width=1.5)
pl.savefig('2D_classprob.png')
p3.remove()
pl.scatter(beachcoor[0],beachcoor[1],marker='o',s=500,facecolors='None',edgecolors='blue',lw=3)
pl.savefig('2D_class.png')
```

```python
bpic.append(np.random.normal(beachcoor[2],3,100))
spic.append(np.random.normal(seacoor[2],10,100))
```
