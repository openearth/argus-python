```python
import os
import numpy as np
import scipy.interpolate
import cPickle as pickle
import pandas as pd

fpath = '../../argusgui/argusgui/datasets/argusnl_new/'

FEATURE_REPLACEMENTS = {'waterpool':'water',
                        'watersea':'water',
                        'objectdune':'object',
                        'objectbeach':'object',
                        'objectsea':'object',
                        'sanddune':'sand',
                        'sandbeach':'sand'}
```

```python
def get_classes(Y, replace=True):
    classes = sorted(np.unique(np.hstack([np.unique(I) for I in Y])))
    if replace:
        return sorted(np.unique(replace_classes(classes)))
    else:
        return classes

def replace_classes(I, replacements=FEATURE_REPLACEMENTS):
    J = np.array(I)
    for k, v in replacements.iteritems():
        J[J == k] = v
    return J
```

```python
# set resulting grid size
n = 100

# determine classes
classes = get_classes(Y_train)

# allocate normalized relation maps
#maps = {c1:{c2:np.zeros((2*n,2*n)) for c2 in classes} for c1 in classes}
maps = pd.Panel4D(labels = classes,items = classes,major_axis = range(2*n), minor_axis = range(2*n))
maps = maps.fillna(0)

# loop over training samples
for k, I in enumerate(Y_train):
    
    I = replace_classes(I)
    
    nx, ny = I.shape

    # allocate unnormalized relation maps
    mapi = {c1:{c2:np.zeros((2*nx,2*ny)) for c2 in classes} for c1 in classes}
    
    # loop over classes in current image
    for c in np.unique(I):
    
        # determine score matrix (1 for superpixel of current class, 0 otherwise)
        scores = (I == c).astype(np.uint)
    
        # loop over all superpixels
        for i in range(nx):
            for j in range(ny):
                
                # determine offset from center
                x1 = nx - i
                y1 = ny - j
                x2 = 2*nx - i
                y2 = 2*ny - j
            
                # determine class of current superpixel
                c0 = I[i,j]
                
                # add score matrix with offset to relation map
                mapi[c][c0][x1:x2,y1:y2] += scores*
        
        # loop over all classes
        for c0 in classes:
            
            # determine x,y axes in pixelspace of relation map
            # start at 0.5 to account for the fact that the first superpixel is 0.5 superpixel width/height from the border
            xn = np.arange(0.5,mapi[c][c0].shape[0],1)
            yn = np.arange(0.5,mapi[c][c0].shape[1],1)
            
            # interpolation object
            f = scipy.interpolate.interp2d(yn,xn,mapi[c][c0])
            
            # determine 2*n equidistant locations in pixelspace
            xi = np.linspace(0,mapi[c][c0].shape[0],2*n)
            yi = np.linspace(0,mapi[c][c0].shape[1],2*n)
            
            # interpolate to 4*n*n equidistant points for normalization
            maps[c][c0] += f(yi,xi)
            
# Add simple uniform prior...
maps += 1
            
# Normalization
tot = maps.sum(axis = 'labels')
for c1 in classes:
    maps.ix[:,c1,:,:] = maps.ix[:,c1,:,:].divide(tot[c1])
maps = maps.fillna(0)
```

```python
def vote_image(Ipred1):
    # Class voting based on 1st order prediction
    nx, ny = Ipred1.shape
    
    votes = pd.Panel(items = classes,major_axis = range(nx),minor_axis = range(ny))
    votes = votes.fillna(0)
    
    xn = np.linspace(.5,199.5,200)/100*nx
    yn = np.linspace(.5,199.5,200)/100*ny
    
    # For every superpixel, loop over the relative location maps
    for i in range(nx):
        for j in range(ny):
            # determine offset from center
            x1 = nx - i - 1
            y1 = ny - j - 1
            x2 = 2*nx - i - 1
            y2 = 2*ny - j - 1
            
            c0 = Ipred1[i,j]
            for c in classes:
                f = scipy.interpolate.interp2d(yn,xn,maps[c][c0].as_matrix())
                votes[c] += f(arange(y1,y2),arange(x1,x2))
                
    Ipred2 = pd.DataFrame([votes.major_xs(i).idxmax(axis=1) for i in range(votes.shape[1])])
    
    return Ipred2.as_matrix(), votes
```
