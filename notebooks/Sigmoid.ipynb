```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu1 = -1
mu2 = 1
sigma1 = .5
sigma2 = 2

r  = 10
x  = np.linspace(-r,r,1000)
w  = abs(mu1-mu2) / (sigma1 * sigma2)

f1 = scipy.stats.norm(mu1, sigma1)
f2 = scipy.stats.norm(mu2, sigma2)
```

```python
fig, axs = plt.subplots(figsize=(10,8))
axs.plot(x, f1.pdf(x), '-b')
axs.plot(x, f2.pdf(x), '-r')
axs.plot(x, f1.pdf(x) + f2.pdf(x), '-g')
axs.set_xlim((-r,r))
```

```python
Z = f1.pdf(x) + f2.pdf(x)

fig, axs = plt.subplots(figsize=(10,8))
axs.plot(x, 1 - np.divide(f1.pdf(x), Z), '-b')
axs.plot(x, np.divide(f2.pdf(x), Z), '--r')
axs.plot(x, 1 / (1 + exp(-w*x)), '--g')
axs.set_xlim((-r,r))
```
