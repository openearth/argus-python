```python
ds = 'argusnl'
images = filesys.get_image_list(ds)
images = [im for im in images if 'kijkduin' in im and 'c2' in im]
images
```

```python
f = filesys.read_feature_files(ds, im, ext='linear')[0]
f2 = filesys.read_feature_files(ds, im, ext='normalized')[0]

c = filesys.read_export_file(ds, im, 'classes')
c = np.asarray([x.replace('beach', '').replace('dune', '').replace('sea', '') for x in c])

f['class'] = c
f2['class'] = c
```

```python
fig, axs = subplots()

features = ['mean_intensity.%d' % i for i in range(3,0,-1)]

f.filter(features).groupby(c).mean().plot(kind='bar', legend=False, ax=axs)

axs.legend(loc='upper left', bbox_to_anchor=(1., 1.))
```

```python
features = ['mean_intensity.%d' % i for i in range(3,0,-1)]
classes = np.unique(c)

fig, axs = subplots(len(classes), 1, figsize=(20,10))

for i, cls in enumerate(classes):
    f.filter(features).iloc[np.where(c==cls)[0][:100]].plot(kind='bar', legend=False, ax=axs[i])
    axs[i].set_title(cls)
#    axs[i].set_xticks(())
    axs[i].set_ylim((0, 255))

fig.tight_layout()
#axs.legend(loc='upper left', bbox_to_anchor=(1., 1.))
```
