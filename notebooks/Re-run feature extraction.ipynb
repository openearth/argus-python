```python
import os
import re
import cPickle as pickle
import numpy as np
import pandas
from matplotlib import pyplot as plt

from argus2 import classification as cls

fpath = '../../argusgui/argusgui/datasets/argusnl_new/'
```

```python
for fname in os.listdir(fpath):
    if re.match('^\d+.+\.jpg$', fname):
        print fname
        
        # determine file names
        imgfile = os.path.join(fpath,'cropped_%s' % fname)
        segfile = os.path.join(fpath,re.sub('\.[^\.]+$','.segments.pkl',fname))
        feafile = os.path.join(fpath,re.sub('\.[^\.]+$','.features.pkl',fname))
        
        # read image
        img = plt.imread(imgfile)
        
        # read segments
        fp = open(segfile,'rb')
        seg = pickle.load(fp)
        fp.close()
        
        # extract and linearize features
        features = cls.classification.extract_features(img, np.array(seg['segments']))
        feature_list, keys_0d = cls.classification.make_features_0d(features)
        
        # write features
        fp = open(feafile,'wb')
        pickle.dump((features,seg['nx'],seg['ny'],feature_list,keys_0d),fp)
        fp.close()
```
