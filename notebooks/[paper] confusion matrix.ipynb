```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cm = np.array([[14890,898,24,1084,385],
               [902,42509,49,274,1690],
               [83,41,14835,8,510],
               [1405,243,12,20786,58],
               [223,1491,284,120,36720]])

cm_n = np.empty(cm.shape)
for i,c in enumerate(cm):
    cm_n[i] = c/np.float(c.sum())*100.
```

```python
cm_n
```

```python
percs_man = np.array([[86.2,5.2,0.1,6.3,2.2],
               [2.0,93.6,0.1,0.6,3.7],
               [.5,.2,95.9,.1,3.3],
               [6.2,1.1,0.1,92.4,0.2],
               [0.6,3.8,0.7,0.3,94.6]])

# Manually tweaked decimals to make rows sum up to 100%
```

```python
m,n = cm.shape
cms = cm.sum(axis=1)
cmnorm = cm.copy().astype(float)
for i in range(m):
    cmnorm[i,:] = cm[i,:]/np.float(cms[i])*100
fig,ax = plt.subplots(figsize=(8,8))
p = ax.matshow(cm_n,cmap='gray_r',vmin=0.,vmax=100.)   #'BrBG')
ax.set_xticklabels(['dummy','object','sand','sky','vegetation','water'],fontsize=15)
ax.set_yticklabels(['dummy','object','sand','sky','vegetation','water'],fontsize=15)
ax.xaxis.set_label_position('top')
ax.set_xlabel('Predicted',fontsize=18)
ax.set_ylabel('Annotated',fontsize=18)
cb = plt.colorbar(p,fraction=0.046, pad=0.04)
cb.set_ticks([0,50,100])
cb.set_ticklabels(['0 %','50 %','100 %'])
cb.ax.tick_params(labelsize=15)

for i in range(m):
    for j in range(n):
        if i == j:
            col = 'white'
        else:
            col = 'black'
        plt.text(j,i,'(%d)' %cm[i,j],horizontalalignment='center',verticalalignment='top',color=col,fontsize=16)
        plt.text(j,i,'%.1f %%' %percs_man[i,j],horizontalalignment='center',verticalalignment='bottom',color=col,fontsize=16)
```
