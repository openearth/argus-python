```python
_FREQS = np.arange(0.05, 0.30, 0.1)
_SIGMAS = [3, 5, 7]
_SIGMA1 = range(1, 21, 7)
_THETAS = np.arange(0, np.pi, 0.25*np.pi)
```

```python
channelstats = [{'min': 0., 'max': []} for i in range(len(_FREQS)*len(_SIGMAS))]
for i,(frequency,sigma) in enumerate(product(_FREQS,_SIGMAS)):
    k = filter.gabor_kernel(frequency=frequency,theta=0, sigma_x=sigma, sigma_y=sigma)
    resp = np.zeros(k.shape) + 255.
    resp_pos = resp.copy()
    resp_neg = resp.copy()
    
    #resp_pos[np.real(k) < -1e-10] = 0.
    #resp_pos[np.imag(k) < -1e-10] = 0.
    resp_pos[np.real(k)+np.imag(k) < 1e-10] = 0.
    resp_real = np.sum(resp_pos*np.real(k))
    resp_imag = np.sum(resp_pos*np.imag(k))
    resp_sgn = np.sign(resp_imag + resp_real)
    resp_pos = resp_sgn * np.sqrt(resp_real**2 + resp_imag**2)
    
    #resp_neg[np.real(k) > 1e-10] = 0.
    #resp_neg[np.imag(k) > 1e-10] = 0.
    resp_neg[np.real(k)+np.imag(k) > -1e-10] = 0.
    resp_real = np.sum(resp_neg*np.real(k))
    resp_imag = np.sum(resp_neg*np.imag(k))
    resp_sgn = np.sign(resp_imag + resp_real)
    resp_neg = resp_sgn * np.sqrt(resp_real**2 + resp_imag**2)
    
    channelstats[i]['max'] = resp_pos
    channelstats[i]['min'] = resp_neg
```
