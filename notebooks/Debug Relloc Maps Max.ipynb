```python
from flamingo import batch, filesys
from flamingo.classification import utils
from flamingo.classification.features import relativelocation
from matplotlib import pyplot as plt
ds = 'argusnl'
```

```python
images = filesys.get_image_list(ds)
clist = filesys.read_default_categories(ds)
maps = filesys.read_export_file(ds,None,'relative_location_maps')
```

```python
agg = {'sand': (u'sandbeach','sanddune'),'object': ('objectbeach','objectdune','objectsea'),'water': ('watersea','waterpool')}
clist = list(np.unique(utils.aggregate_classes(np.array(clist),agg)))
```

```python
n = 20
images = images[:n]
fig, ax = plt.subplots(nrows=len(images), ncols=7, figsize=(35,10 * round(n / 2)))

for i,im in enumerate(images):
    classes = np.array(filesys.read_export_file(ds,im,'classes'))
    classes = utils.aggregate_classes(classes,agg)
    meta = filesys.read_export_file(ds,im,'meta')
    segments = filesys.read_export_file(ds,im,'segments')
    
    clnum = np.empty(meta['image_resolution_cropped'][0:2])
    for j in range(segments.max()+1):
        clnum[segments == j] = clist.index(classes[j])
        
    votes,Ivote = relativelocation.vote_image(classes.reshape(meta['superpixel_grid']),maps)
    
    votnum = np.empty(meta['image_resolution_cropped'][0:2])
    for j in range(segments.max()+1):
        votnum[segments == j] = clist.index(Ivote.ravel()[j])
        
    ax[i,0].imshow(clnum,clim=(0,4));
    ax[i,0].set_title('Annotation')
    ax[i,1].imshow(votnum,clim=(0,4));
    ax[i,1].set_title('Voted image')
    ax[i,2].imshow(votes['sky'],clim=(0,1));
    ax[i,2].set_title('P(sky)')
    ax[i,3].imshow(votes['water'],clim=(0,1));
    ax[i,3].set_title('P(water)')
    ax[i,4].imshow(votes['sand'],clim=(0,1));
    ax[i,4].set_title('P(sand)')
    ax[i,5].imshow(votes['vegetation'],clim=(0,1));
    ax[i,5].set_title('P(vegetation)')
    ax[i,6].imshow(votes['object'],clim=(0,1));
    ax[i,6].set_title('P(object)')
```

```python
def cl2num(im,cl):
    num = np.empty(shape(im))
    for i,c in enumerate(cl):
        num[im == c] = i
    
    return num
```

```python
mws = np.empty((200,200))
mww = mws.copy()
msw = mws.copy()
mss = mws.copy()

mws[:,:100] = np.ones((200,100))
mws[:,100:] = np.zeros((200,100))

mww[:,:100] = np.ones((200,100))
mww[:,100:] = np.zeros((200,100))

msw[:,100:] = np.ones((200,100))
msw[:,:100] = np.zeros((200,100))

mss[:,100:] = np.ones((200,100))
mss[:,:100] = np.zeros((200,100))

map_art = {'water':{'water':mww, 'sand':mws},'sand':{'water':msw,'sand':mss}}

cl_art = np.empty(shape=(22,27),dtype='<U5')
cl_art[:,:14] = u'water'
cl_art[:,14:] = u'sand'
```

```python
votes_art,Ivote_art = relativelocation.vote_image(cl_art,map_art)
```

```python
iv_art_num = cl2num(Ivote_art,['water','sand'])
cl_art_num = cl2num(cl_art,['water','sand'])
```
